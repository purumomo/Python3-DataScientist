```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = [1,2,3]
b = [4,5,6]
plt.plot(a,b)
plt.show
```

```python
plt.plot(a,b, '--') #Line
```

```python
plt.plot(a,b, '*') #Line
```

```python
c =[10,8,6]
d =[1,4,2]
plt.plot(a,b,'--',c,d,'*')
```

```python
t = np.arange(0.0,2.0,0.1)
s = np.sin(t*np.pi)
plt.plot(t,s,'r--',label='aaaaa')
plt.plot(t*2,s,'b--',label='bbbb')
plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('this is a demo')
plt.legend()
```
