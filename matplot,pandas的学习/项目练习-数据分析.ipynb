```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from datetime import datetime


```

```python
start = datetime(2015,9,20)
alibaba = pd.read_csv('BABA.csv')
amazon = pd.read_csv('AMZN.csv')
```

```python
amazon.head()
```

```python
alibaba.head()
```

```python
alibaba['Adj Close'].plot(legend=True)
amazon['Adj Close'].plot(legend=True)
```

```python
alibaba['Volume'].plot(legend=True)
amazon['Volume'].plot(legend=True)
```

```python
alibaba['Hight-Low'] = alibaba['High'] - alibaba['Low']
amazon['Hight-Low'] = amazon['High'] - amazon['Low']
```

```python
alibaba['daily-return'] = alibaba['Adj Close'].pct_change()
alibaba['daily-return'].plot(figsize=(10,4),linestyle='--',marker='o')
```
