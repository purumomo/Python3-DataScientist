```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0,5.0)
x
```

```python
len(x)
```

```python
#
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*2*x)
plt.plot(x,y1,'b--',label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x,y2,'r--',label='sin(pi*x*2)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend()
```

```python
plt.show()
```

```python
#subplot   
plt.subplot(221)
plt.plot(x,y1,'b--')
plt.subplot(222)
plt.plot(x,y2,'r--')
plt.subplot(223)
plt.plot(x,y1,'r*')
plt.subplot(224)
plt.plot(x,y2,'b*')
```

```python
figure,ax =plt.subplots(2,2) 
#ax array
ax[0][0].plot(x,y1)
ax[0][1].plot(x,y2)
ax[1][0].plot(x, y1,'r--')
ax[1][1].plot(x, y1,'b--')
plt.show()
```
