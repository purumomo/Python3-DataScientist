```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# 返回0到1之间的随即小数
a = np.random.rand()
print(a)
# 均匀随机数
```

```python
# 均匀随机数的散点图分布
n = 800
x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
#偏移随机数
n = 800
a = np.random.randn(n)
b = np.random.randn(n)
plt.scatter(a,b)
plt.grid()
plt.show()
```
