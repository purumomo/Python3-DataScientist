```python
import matplotlib.pyplot as plt
import numpy as np

#梯度下降
# 原函数
def f(x):
    return x**2
#原函数的梯度
def g(x):
    return 2*x
# 学习率
eta = 0.1
# 初始值
x = 10
# 存储值
values = []
#迭代20次
for i in range(20):
    x = x - eta * g(x)
    values.append(f(x))

    
    
plt.plot(values
plt.xlabel = ('iteration')
plt.xlabel = ('function value')
plt.title('Changes in function values ​​during gradient descent')
plt.show()
```
