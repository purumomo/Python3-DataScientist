```python
#概率密度函数
import numpy as np
import matplotlib.pyplot as plt

def pdf(x,mu,sigma):
    return 1/(sigma*np.sqrt(2*np.pi)*np.exp(-(x-mu)**2/(2*sigma**2)))

x = np.linspace(-5,5)
y_1=pdf(x,0,0.5) #平均值维0，标准差为0.5
y_2=pdf(x,0,1.0) #平均值维0，标准差为1.0
y_3=pdf(x,0,2.0) #平均值维0，标准差为2.0

plt.plot(x,y_1,label="σ:0.5",linestyle="dashed")
plt.plot(x,y_2,label="σ:1.5",linestyle="solid")
plt.plot(x,y_3,label="σ:2.0",linestyle="dashdot")
plt.legend()
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成随机数据
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1000)

# 绘制直方图
count, bins, ignored = plt.hist(s, 30, density=True)

# 绘制概率密度函数
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
         np.exp(- (bins - mu)**2 / (2 * sigma**2)),
         linewidth=2, color='r')

plt.show()
```
