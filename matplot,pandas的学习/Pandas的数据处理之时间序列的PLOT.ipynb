```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
t_range = pd.date_range('2016-01-01','2016-12-31')
t_range
```

```python
s = Series(np.random.rand(len(t_range)),index=t_range)
s
```

```python
s['2016-01'].mean() #求平均值
```

```python
s_month = s.resample('M').mean()
s_month
```

```python
s_hour = s.resample('M').bfill()
s_hour
```
