```python
#查询数据表的key
df.index
```

```python
#创建一个新的df
df_new = DataFrame(df, columns=['Python page' , 'Python'])
df_new
```

```python
#创建一个新的列 并且赋值 可以用Series进行赋值
df_new = DataFrame(df,columns=['Python page' , 'Python','testColum'])
df_new['testColum'] = Series(range(0,10))
df_new
```

```python
#如何给指定的Index赋值 用Series来给Key 和 Values 赋值
df_new = DataFrame(df,columns=['Python page' , 'Python','test2022'])
df_new['test2022'] = Series([100,200],index=[2,4])
df_new
```
