```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot():
    x = np.linspace(0,14,100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x,np.sin(x+1)*(x+0.75),label='sin(x+i)*(i+0.75)')
    plt.legend()
```

```python
sinplot()
```

```python
import seaborn as sns
sinplot()
```

```python
with sns.axes_style('white'):
    sinplot()
```
